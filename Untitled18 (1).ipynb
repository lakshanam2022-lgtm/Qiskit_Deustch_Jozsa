```python
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator

# --- Adaptive balanced oracle (works for any n ≥ 2) ---
def oracle_balanced_custom(qc, qubits, ancilla):
    """
    Flexible balanced oracle:
    Flips the ancilla for half of all possible inputs.
    Automatically adapts to the number of input qubits.
    """
    if len(qubits) == 1:
        qc.cx(qubits[0], ancilla)
    elif len(qubits) == 2:
        qc.cx(qubits[0], ancilla)
        qc.cx(qubits[1], ancilla)
    elif len(qubits) == 3:
        qc.cx(qubits[0], ancilla)
        qc.ccx(qubits[1], qubits[2], ancilla)
    else:
        # For n ≥ 4, flip ancilla for alternate qubits
        for i in range(0, len(qubits), 2):
            qc.cx(qubits[i], ancilla)
    qc.barrier()
    return qc


# --- Deutsch–Jozsa circuit builder ---
def deutsch_jozsa_circuit(n, oracle_func):
    qc = QuantumCircuit(n + 1, n)
    qubits = list(range(n))
    ancilla = n

    # Initialize ancilla to |1>
    qc.x(ancilla)
    qc.h(ancilla)

    # Create superposition on input qubits
    qc.h(qubits)

    # Apply oracle
    oracle_func(qc, qubits, ancilla)

    # Apply Hadamard again on input qubits
    qc.h(qubits)
    qc.barrier()

    # Measure
    qc.measure(qubits, range(n))
    return qc

def run_dj(qc, desc=""):
    sim = AerSimulator()
    compiled = transpile(qc, sim)
    result = sim.run(compiled, shots=1024).result()
    counts = result.get_counts()
    print(f"\n=== {desc} ===")
    display(qc.draw("text"))  # 👈 changed from 'mpl' to 'text'
    print("Measurement counts:", counts)
    plot_histogram(counts)
    return counts

# --- Test for different numbers of qubits ---
for n in [2, 4, 5]:
    qc = deutsch_jozsa_circuit(n, oracle_balanced_custom)
    run_dj(qc, f"Deutsch–Jozsa with n = {n}")
```

```python
# =========================================================
# 💡 TASK 4 — Deutsch–Jozsa Algorithm (Qiskit 2.x+ Compatible)
# =========================================================

# --- Install all required libraries ---
!pip install qiskit qiskit-aer matplotlib pylatexenc --quiet

# --- Imports ---
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# =========================================================
# 🧩 Step 1 — Define Oracles
# =========================================================

def oracle_constant(qc, n):
    """Oracle that gives a constant output (always 0)."""
    qc.barrier()

def oracle_balanced(qc, n):
    """Oracle that flips output for half of the inputs (balanced)."""
    qc.cx(0, n)
    qc.cx(1, n)
    qc.barrier()

# =========================================================
# ⚙️ Step 2 — Build Deutsch–Jozsa Circuit
# =========================================================

def deutsch_jozsa_circuit(n, oracle_func):
    """Build Deutsch–Jozsa circuit for n input qubits."""
    qc = QuantumCircuit(n + 1, n)
    qc.x(n)                      # Set output qubit to |1>
    qc.h(range(n + 1))           # Apply Hadamard to all qubits
    oracle_func(qc, n)           # Apply the oracle
    qc.h(range(n))               # Apply Hadamard to input qubits
    qc.measure(range(n), range(n))
    return qc

# =========================================================
# 🧠 Step 3 — Create Circuit
# =========================================================

n = 3
qc_balanced = deutsch_jozsa_circuit(n, oracle_balanced)

print("\n✅ Deutsch–Jozsa Circuit (Balanced Oracle):")
try:
    qc_balanced.draw('mpl')
except Exception as e:
    print("⚠️ Could not render using Matplotlib. Showing text mode instead:")
    print(qc_balanced.draw('text'))

# =========================================================
# 🧪 Step 4 — Simulate using QASM Simulator
# =========================================================

backend = Aer.get_backend('qasm_simulator')

# --- Run balanced oracle ---
tqc_bal = transpile(qc_balanced, backend)
result_bal = backend.run(tqc_bal, shots=1024).result()
counts_balanced = result_bal.get_counts()

# --- Run constant oracle ---
qc_constant = deutsch_jozsa_circuit(n, oracle_constant)
tqc_con = transpile(qc_constant, backend)
result_con = backend.run(tqc_con, shots=1024).result()
counts_constant = result_con.get_counts()

# =========================================================
# 📊 Step 5 — Display Results
# =========================================================

print("\n🧩 Balanced Oracle Measurement Results:")
print(counts_balanced)

print("\n🧩 Constant Oracle Measurement Results:")
print(counts_constant)

plot_histogram(
    [counts_constant, counts_balanced],
    legend=['Constant Oracle', 'Balanced Oracle']
)
plt.show()
```
